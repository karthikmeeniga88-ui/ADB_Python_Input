```python
df=spark.read.option('header',True).csv('/Volumes/workspace/default/input/Customer.csv')
df.display()
```

```python
from pyspark.sql.functions import col
df1=df.withColumn("CustomerID",col('CustomerID').cast('integer'))\
      .withColumn("CustomerCreditLimit",col('CustomerCreditLimit').cast('integer'))
display(df1)
```

```python
df1.select ('CustomerID','CustomerName','CustomerEmail','CustomerAddress','CustomerCreditLimit').display()
```

```python
df2=df1.na.drop(subset='CustomerName')
df2.display()
```

```python
df2.write.format('delta').mode('overwrite').save('/Volumes/workspace/default/input/Retail_customer')
#df3=df2.drop('CustomerPhone')
#df3.display()
```
